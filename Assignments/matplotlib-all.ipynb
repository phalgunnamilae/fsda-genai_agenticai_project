```python
x=np.linspace(0,2,100)
```

```python
plt.plot(x,x,label='linear')
plt.plot(x,x*2,label='quadratic')
plt.plot(x,x**2, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title('Simple Graph')

plt.legend()
plt.show()
```

```python
fig = plt.figure()
x2 = np.linspace(0,5,10)
y2 = x2**2
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x2,y2,'r',ls=":",marker='o')

axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('Simple Graph')
```

```python
fig = plt.figure()
x2 = np.linspace(0,5,10)
y2 = x2**2
axes = fig.add_axes([10,0,0.8,0.8])
axes.plot(x2,y2,'r',ls=":",marker='o')

axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('Simple Graph')
```

```python
#multiline plots

x4 = range(1,5)
plt.plot(x4, [xi*1.5 for xi in x4], label = 'first plot')
plt.plot(x4, [xi*3 for xi in x4], label = 'second plot')
plt.plot(x4, [xi/3.0 for xi in x4], label = 'third plot')
plt.legend()
plt.show()
```

```python
#scatter plot

x7 = np.linspace(0,10,30)
y7 = np.sin(x7)

plt.scatter(x7, y7)
```

```python
#Histogram

data1 = np.random.randn(1000)
print(data1)

plt.hist(data1); 
```

```python
#bar chart

data2 = [5,25,50,20]
plt.bar(range(len(data2)), data2)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Bar chart')
plt.show()
```

```python
#Horizontal Bar chart

data3 = [5,10,15,20,14,9]
plt.barh(range(len(data3)), data3)
plt.show()
```

```python
# Pie chart

fig = plt.figure()
y10 = ['Computer', 'Electornics' , 'Electrical', 'Mechanical']
x10 = [45,25,20,10]

plt.pie(x10, labels=y10)
plt.show()
```

```python
#Area chart

x11 = [1,4,3,7]
y11 = [4,6,3,2]
plt.fill_between(x11,y11)
plt.show()
```
