```python
for i in titanic_dataset.select_dtypes(include=['number']).columns :
    titanic_dataset[i].fillna(titanic_dataset[i].median(), inplace=True)
```

```python
for i in titanic_dataset.select_dtypes(include=['object']).columns :
    titanic_dataset[i].fillna(titanic_dataset[i].mode()[0], inplace=True)
```
